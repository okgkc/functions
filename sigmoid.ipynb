```python
import matplotlib.pyplot as plt
import numpy as np
import math
  
x = np.linspace(-7, 7, 1000)
z = 1/(1 + np.exp(-x))
  
plt.plot(x, z, color= "r")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.title("Sigmoid Function", fontsize = 14)

plt.text(-6.5, 0.88, 'y = 1 / (1+ exp(-x)) ', fontsize = 13,  bbox = dict(facecolor = 'white', alpha = 0.5))

plt.grid()

  
plt.show()
```
